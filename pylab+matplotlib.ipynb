```python
import matplotlib.pyplot as plt
import pylab as pl
```

```python
x = range(10)  # 横轴的数据
y = [i*i for i in x]  # 纵轴的数据
pl.plot(x, y,'og-.')  # 调用pylab的plot函数绘制曲线
pl.xlabel(u"我是横轴")
pl.ylabel(u"eee")
pl.show()
```

```python
pl.plot(x, y, 'ob-', label=u"y=x^2曲线图")  # 加上label参数添加图例
pl.legend()  # 让图例生效
pl.show()
```
